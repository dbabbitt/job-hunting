```python

%%time
import re
import random
import textwrap
from neo4j import GraphDatabase

ZMQInteractiveShell_obj = get_ipython()
uri = 'bolt://localhost:7687'
user = 'neo4j'
password = 'Genesis11'
driver = GraphDatabase.driver(uri, auth=(user, password))
```

```python

import pandas as pd

def do_cypher_tx(tx, cypher):
    results_list = tx.run(cypher)
    values_list = []
    for record in results_list:
        values_list.append(dict(record.items()))
    
    return values_list
```

```python

# Get the HTML strings that have "BA" or "BS" in them
tx_str = r"""
MATCH (np:NavigableParents)
WHERE np.navigable_parent =~ '.*\\b(B[AS]|B\\.[AS]\\.)\\b.*'
SET np.is_educational_requirement = 1
RETURN PROPERTIES(np) AS properties;"""
with driver.session() as session:
    rows_list = [row_dict['properties'] for row_dict in session.write_transaction(do_cypher_tx, tx_str)]
for navigable_parent in pd.DataFrame(rows_list).navigable_parent.tolist():
    print(navigable_parent[:100])
```
